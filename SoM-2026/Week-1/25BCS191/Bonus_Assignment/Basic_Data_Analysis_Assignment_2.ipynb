### 7) Which was the most-ordered item?

```python
#your code here
most_ordered_item = chipo['item_name'].value_counts().idxmax()
most_ordered_item
```

### 8) For the most-ordered item, how many items were ordered?

```python
#your code here
cond=chipo['item_name']==most_ordered_item
chipo.loc[cond,'quantity'].sum()
```

### 9) What was the most ordered item in the choice_description column?

```python
#your code here
most_ordered_item2 = chipo['choice_description'].value_counts().idxmax()
most_ordered_item2
```

### 10) How many items were ordered in total?

```python
#your code here
total = chipo['quantity'].sum()
total
```

### 11) Turn the item price into a float

```python
#your code here
chipo['item_price']=chipo['item_price'].str.replace('$','',regex=False).astype(float)
```

### 12) How much was the revenue for the period in the dataset?
#### Revenue will be the sum of quantity * item_price!

```python
#your code here
revenue=((chipo['quantity'])*(chipo['item_price'])).sum()
revenue
```

### 13) How many orders were made in the period?

```python
#your code here
no_of_orders=chipo['order_id'].unique().shape[0]
no_of_orders
```

### 14) What is the average revenue amount per order?

```python
#your code here
avg_rev=revenue/no_of_orders
avg_rev
```

### 16) Show a histogram with number of orders with bill on the x-axis.

```python
import seaborn as sns
from matplotlib import pyplot as plt
bill=((chipo['quantity'])*(chipo['item_price']))
bins = np.arange(0, 151, 5)
sns.histplot(data=chipo, x=bill, bins=bins)
plt.xlabel('Bill')
```
