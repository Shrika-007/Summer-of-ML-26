### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python

type_totals = df.groupby('Type 1')['Total'].mean().reset_index()


type_totals = type_totals.sort_values('Total', ascending=False)


sns.set(style="whitegrid")


plt.figure(figsize=(12, 6))
ax = sns.barplot(data=type_totals, x='Type 1', y='Total', palette='Set2', hue='Type 1', legend=False)


plt.title('Average Total Stats by Primary Pokémon Type')
plt.xlabel('Primary Type')
plt.ylabel('Average Total Stat')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
ax = sns.scatterplot(
    data=df,
    x='Speed',
    y='Total',
    hue='Legendary',
    palette='Set1',
    alpha=0.7
)


plt.title('Pokémon: Total Stats vs. Speed')
plt.xlabel('Speed')
plt.ylabel('Total Stats')
plt.legend(title='Legendary')
plt.tight_layout()
plt.show()
```

```python
# Calculate Generation-wise average for Total stats
gen_avg = df.groupby('Generation')['Total'].mean().reset_index()



sns.set(style="whitegrid")


plt.figure(figsize=(10, 6))
sns.lineplot(data=gen_avg, x='Generation', y='Total', marker='o', color='mediumseagreen')

plt.title('Average Total Stats by Pokémon Generation')
plt.xlabel('Generation')
plt.ylabel('Average Total Stats')
plt.xticks(gen_avg['Generation'])
plt.tight_layout()
plt.show()
```

```python
#your code here

#plot number 1 count plot 
sns.set_style('darkgrid')
fig, axes = plt.subplots(1,2, figsize=(16,6))

sns.countplot(data=df, y='Type 1', order=df['Type 1'].value_counts().index, palette='Set2', hue='Type 1', 
              legend=False, edgecolor='black', lw=1.5, ax=axes[0])
axes[0].set_title('Type 1 Distribution', fontweight='bold')

sns.countplot(data=df, y='Type 2', order=df['Type 2'].value_counts().index, palette='Set2', hue='Type 2', 
              legend=False, edgecolor='black', lw=1.5, ax=axes[1])
axes[1].set_title('Type 2 Distribution', fontweight='bold')

plt.tight_layout()
```

```python
#your code here

#plot number 2 hist plot,scatterplot nad heatmap
fig = plt.figure(figsize=(20, 12))
axes = [fig.add_subplot(2, 2, 1),  
        fig.add_subplot(2, 2, 2),  
        fig.add_subplot(2, 1, 2)]
sns.histplot(data=df, x='Total', hue='Legendary', kde=True, element='step', palette='Set2', linewidth=1.5,ax=axes[0] )
axes[0].set_title('Distribution of Total Stats (Legendary vs Non-Legendary)', fontweight='bold')

sns.scatterplot(data=df, x='Sp. Atk', y='Sp. Def', hue='Legendary', style='Legendary',  palette='Set2', s=40, ax=axes[1])
axes[1].set_title('Sp. Atk vs Sp. Def (Legendary vs Non-Legendary)' , fontweight='bold')

sns.heatmap(df.corr(numeric_only=True), cmap='RdBu', annot=True, fmt='.2f', linewidths=0.5, vmin=-1, vmax=1, ax=axes[2])
axes[2].set_title('Correlation Heatmap of Pokemon Stats' , fontweight='bold')

plt.tight_layout()
```

```python
#your code here
fig, axes=plt.subplots(1,2, figsize=(16,6))

sns.boxplot(data=df, x='Generation', y='Attack', hue='Legendary', palette='Set2', linewidth=1.5, ax=axes[0])
axes[0].set_title('Attack Distribution by Generation (Legendary vs Non-Legendary)',fontweight='bold')

sns.boxplot(data=df, x='Generation', y='Defense', hue='Legendary', palette='Set2', linewidth=1.5, ax=axes[1])
axes[1].set_title('Defense Distribution by Generation (Legendary vs Non-Legendary)',fontweight='bold')

plt.tight_layout()
```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
non_legendary_pokemons=df.loc[df['Legendary']==False]

Attack_of_Charizard=df.loc['Charizard','Attack']
Defence_of_Charizard=df.loc['Charizard','Defense']

condition=(non_legendary_pokemons['Attack']<Attack_of_Charizard)&(non_legendary_pokemons['Defense']>Defence_of_Charizard)

final_count=non_legendary_pokemons.loc[condition].shape[0]
final_count
```

### Which type of legendary pokemons are the most common?

```python
#your code here
type1_legen = df[['Type 1', 'Legendary']].rename(columns={'Type 1': 'Type'})
type2_legen = df[['Type 2', 'Legendary']].rename(columns={'Type 2': 'Type'})

combined = pd.concat([type1_legen, type2_legen]).dropna(subset=['Type'])
combined.loc[combined['Legendary']==True].sort_values(by='Type').value_counts().idxmax()
```

### How many generation-3, non-legendary pokemons have higher HP than the weakest generation-6, legendary pokemon?

```python
#your code here
parameters_1 = (df['Generation']==6) & (df['Legendary'] == True)
gen6_legen=df.loc[parameters_1].sort_values(by='HP')
weekest_gen6_legen=gen6_legen['HP'].min()

parameters_2 = (df['Generation']==3) & (df['Legendary'] == False) & (df['HP']>weekest_gen6_legen)
gen3_nonlegen=len(df.loc[parameters_2])
gen3_nonlegen
```
