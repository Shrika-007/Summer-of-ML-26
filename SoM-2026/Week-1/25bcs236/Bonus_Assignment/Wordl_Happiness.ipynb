*Task 2*  
1. Display the data of top 5 east asian countries based on 'Generosity'  
2. Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries (based on generosity)  
3. Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 
'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to 
show the relation these countries vs their 'Logged GDP per capita' 

```python
# Display the data of top 5 east asian countries based on 'Generosity' 

#first, we will extract the required data
d = df1[df1['Regional indicator'] == 'East Asia'].sort_values('Generosity', ascending = False).head()
pd.set_option('expand_frame_repr', False)
print(d)

#now, we will visualise it using a barplot
plt.figure(figsize = (12, 5))
sns.barplot(data = d,
            x = 'Country name',
            y = 'Generosity',
            hue = 'Country name',
            palette = 'Set2',
            legend = False)

plt.title('Top 5 east asian countries')
plt.show()

#
```

```python
#Build a plot (line plot) that shows the variation of 'Ladder score' among the above 5 east asian countries 
#(based on generosity)  

plt.figure(figsize = (10, 6))
sns.lineplot(data = d,
             x = 'Country name',
             y = 'Ladder score',
             marker = 'o'
             )

plt.title('Ladder score of top 5 east asian countries')
plt.show()
```

```python
#Create a dataframe object : df_2021 with the following countries: 'China', 'Nepal', 
#'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Now, build a scatter plot to 
#show the relation these countries vs their 'Logged GDP per capita' 

country = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].isin(country)]

plt.figure(figsize = (10, 6))
sns.scatterplot(data = df_2021,
                x = 'Country name',
                y = 'Logged GDP per capita'
                )

plt.title('df_2021 countries vs Logged GDP per capita')
plt.show()
```
