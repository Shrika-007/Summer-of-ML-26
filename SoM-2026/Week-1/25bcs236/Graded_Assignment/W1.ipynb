```python

df['Type 1 + Type 2'] = df['Type 1'] + '+' + df['Type 2']          #creating a new column such that it contains type1+type2
print(df['Type 1 + Type 2'])                                       #verifying our result

print('-------------------------------------------------------')

#Method 1
#We use the concept of list comprehensions and the apply() function, where we check whether the given 
#entry satisfies the condition or not.. if it doesn't it returns 0, and thus, summing the elements of the
#list gives us the number of pokemons that have type1+type2=Psychic+Flying
l1 = [i for i in df['Type 1 + Type 2'].apply(lambda x: 1 if x == 'Psychic+Flying' else 0)]
print('Method-1: ', sum(l1))

#Concise but similar logic, Method 2
#Instead of creating a list, we apply condition on the newly created column and check if it is satisfied
#the return values are again 0 and 1, so the sum will give us the count of the reqd pokemons
l2 = (df['Type 1 + Type 2'] == 'Psychic+Flying').sum()
print('Method-2: ', l2)
```

### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
#bar graph
#aim: to analyse whether a generation is more offensive or defensive using attack / defense ratio
df['ratio'] = df['Attack'] / df['Defense']
plt.figure(figsize = (10, 6))
t = sns.barplot(data = df,
            x = 'Generation',
            y = 'ratio',
            errorbar = None
            )

plt.title('Attack / Defense ratio by generation')
plt.xlabel('Generation')
plt.ylabel('Attack / Defense ratio')
t.bar_label(t.containers[0])

plt.show()

#from the bar, we can see that the most offensive pokemons are from gen3, while the most defensive ones are from gen6
```

```python
#violin plot
#aim: to compare the mobility (Speed) of different pokemons based on their type1 (since type2 has multiple NaN values)

plt.figure(figsize = (15, 5))
sns.violinplot(data = df,
               x = 'Type 1',
               y = 'Speed',
               palette = 'muted',
               hue = 'Type 1',
               legend = False,
               inner = 'quartile'
               )


plt.title('Speed distribution by type 1')
plt.show()

#from this graph, multiple things can be observed for each type-1
#let's take the case of grass. Here, we can see that the distribution is the widest around the median, and it keeps on shrinking around the top and the bottom
#this means that most of the pokemons in the grass type have their speeds around the median, while very less pokemons are very fast, or very slow
#we can use this plot to compare the mobilities of any two types and see which of them is more mobile than the other by checking where the 
#majority of the distribution (pokemons) lie by observing where the violin plot is the widest. 
```

```python
#boxplot
#aim: to check whether legendary pokemons are statistically stronger or not by comparing their Total values
plt.figure(figsize = (10,6))

sns.boxplot(data = df,
            x = 'Legendary',
            y = 'Total',
            hue = 'Legendary',
            palette = 'muted',
            legend = False
            )

plt.title('Legendary vs Non-Legendary')
plt.show()

#from this boxplot, we can clearly see that most of the legendary pokemons lie above the median value of their total scores,
#which is quite different from the distribution of the non-legendary pokemons, where majority of their pokemons lie below the median of total
#also, the entire legendary pokemon's distribution is placed above that of the non-legendary pokemons
#therefore, we can very safely say that legendary pokemons are indeed stronger than the non-legendary ones.
```

### Print out the third slowest pokemon(s) in the dataset.

```python
#3rd slowest when we consider unique speeds (i,e. if there are 2 pokemons with speed 10, 1 with 12 and 4 with 15, 
# we will print 4 pokemons with speed 12
df = df.sort_values('Speed', ascending = True)
speed = df.iloc[0, df.columns.get_loc('Speed')]     #get_loc is used to get the index of the Speed column
count = 0

for i in df['Speed']:           #we want to find out the unique value of the 3rd slowest speed, so we access the values of speed
    if i > speed:               #right now, speed has the min value of speed in the table, and we compare it against all the values
        count += 1              #if this value (i) is > speed, this means we have found a unique value of speed, and this is the 2nd slowest
        speed = i               #storing the new value to compare against
    if count == 2:              #count must be 2, because we want to find the 3rd slowest (0 = slowest, 1 = 2nd slowest)
        break                   #end the loop as soon as we find the 3rd slowest speed

pd.set_option('expand_frame_repr', False)               #to wrap the column in one line
print(df[df.Speed == speed])                            #printing ALL the pokemons with speed = 3rd slowest speed
print('\n')

#3rd slowest just by sorting and printing whosoever ranks 3rd (ascending)
print(df.iloc[2])
```
