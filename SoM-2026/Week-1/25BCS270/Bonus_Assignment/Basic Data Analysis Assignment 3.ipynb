# Task 2: Visualization of Results using Matplotlib
### 2.1 Top 5 East Asian Countries by Generosity
Display the data of the top 5 East Asian countries based on 'Generosity'.

```python
df2 = df1[df1['Regional indicator'] == 'East Asia'].sort_values(by='Generosity', ascending=False).head(5)
```

### 2.2 Line Plot: Ladder Score Variation
Build a line plot that shows the variation of 'Ladder score' among the top 5 East Asian countries (based on generosity) found in the previous step.

```python
fig , ax = plt.subplots(figsize = (10, 6))
x = np.arange(len(df2.index))
ax.plot((df2['Country name'].values), df2['Ladder score'], 'ro--')
ax.set(xlabel='Country Names', ylabel='Label Score', title='Ladder Score Variation \nof top Five Generous Countries', yticks=df2['Ladder score'])
ax.grid(axis='y', alpha = 0.5)

plt.tight_layout()
plt.show()
fig.savefig('assignment-3-1.png', dpi = 300, bbox_inches = 'tight', pad_inches = 1)
```

### 2.3 Scatter Plot: Logged GDP per Capita
Create a dataframe object named `df_2021` with the following countries: 'China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan'. Build a scatter plot to show the relationship between these countries and their 'Logged GDP per capita'.

```python
l = ['China', 'Nepal', 'Bangladesh', 'Pakistan', 'Myanmar', 'India', 'Afghanistan']
df_2021 = df1[df1['Country name'].apply(lambda x: x in l)]
fig , ax = plt.subplots(figsize = (10,6))
ax.scatter(x=df_2021['Country name'], y= df_2021['Logged GDP per capita']) 

ax.set(xlabel='Country Names', ylabel='Logged GDP per capita', title='Scatter Plot: Logged GDP per Capita')
ax.grid(alpha = 0.5)

plt.tight_layout()
plt.show()
fig.savefig('assignment-3-2.png', dpi = 300, bbox_inches = 'tight', pad_inches = 1)
```
