### 13) What is the mean age of users?

```python
#your code here
print("The mean age of the users is :",users['age'].mean())
```

### 14) What is the age with least occurrence?

```python
#your code here
a = (users['age'].value_counts())[users['age'].value_counts() == users['age'].value_counts().sort_values().values[0]].index
a = [str(x) for x in a.values]
print("The least occuring ages are :")
print(', '.join(a))
```

### 15) What is the zip code with most occurrence?

```python
#your code here
b = (users['zip_code'].value_counts())[users['zip_code'].value_counts() == users['zip_code'].value_counts().sort_values(ascending=False).values[0]].index
print("The most occuring Zip code is :", ', '.join(b.values))
```

### 16) Which profession has youngest people(apart from student)? Show the trend of mean age of people with profession as the seperating factor.

```python
#your code here
g = users.groupby('occupation')
print("The trend of mean age of people with profession :")
print(g['age'].mean().sort_values())
print(f"The occupation which has the second lowest mean age is : {g['age'].mean().sort_values().index[1]}\nwhich means the state of being unemployed.")
```
