### 6. Using seaborn make different types of plots, observe the trend and answer the questions given in the form.

```python
#your code here
#print(df.dtypes) #The columns doesn't require int64 dtype and category reduces memory usage.
#print(df.describe()) ##Analysing the min and max values, and knowing that there are no negative numbers and many are less than 255 values.
dff = pd.read_csv("W1_Pokemon.csv", dtype={
'#'             :'uint16',
'Type 1'        :'category',
'Type 2'        :'category',
'Total'         : 'uint16',
'HP'            : 'uint16',
'Attack'        : 'uint8',
'Defense'       : 'uint8',
'Sp. Atk'       : 'uint8',
'Sp. Def'       : 'uint8',
'Speed'         : 'uint8',
'Generation'    : 'uint8'
})
dff.drop('#', axis = 1, inplace = True)

dff['types'] = [y +'+' + x if (type(x) == str ) else y for x, y in zip(dff['Type 2'], dff['Type 1'])]
dff['element'] = ['Double' if (type(x) == str ) else 'Single' for x in dff['Type 2']]
dff['mega'] = (dff['Name'].apply(lambda x: 'Mega' in x))

sns.set_style('white')

fig, axi = plt.subplots(nrows = 2, ncols = 2, figsize = (35, 25))
sns.scatterplot(data = dff,x = dff.index , y ='Total', hue='Legendary', style='Generation', size = 'mega', alpha = 0.7, ax=axi[0][0], sizes = [50, 150])
axi[0][0].set_title('Scatterplot representing the total power of the pokemons\n w.r.t "gen" and "legendary"',fontsize=15)
axi[0][0].set_xlabel('Index(pokemon list)', fontsize=15)
axi[0][0].set_ylabel('Total power level', fontsize=15)
axi[0][0].tick_params(axis='both', labelsize=10)
axi[0][0].grid(axis = 'y', alpha = 0.4)
axi[0][0].legend(loc='upper left', framealpha = 0.4, title= 'Categories:-')
#FROM THE 1ST PLOT WE CAN CLEARLY SEE THE LEGENDARY POKEMON HAVE HIGHER POWER LEVEL AND THE 4th GEN legendary pokemons are more powerful than others, other MEGA  non legendary pokemons are rivaling the Legendary in terms of power.

sns.boxplot(data= dff, x= 'Generation', y='Total', hue='Legendary', ax=axi[0][1])
axi[0][1].set_title('Boxplot showing the TOTAL POWER of pokemon\n vs.\nGeneration , Legendary',fontsize=15)
axi[0][1].set_xlabel('Generation of pokemon', fontsize=15)
axi[0][1].set_ylabel('Total power', fontsize=15)
axi[0][1].tick_params(axis='both', labelsize=10)
axi[0][1].grid(axis = 'y', alpha = 0.4)
axi[0][1].legend(loc='upper right', framealpha= 0.4, title= 'Legendary :-')
sns.boxplot(data = dff, x = 'mega', y = 'Total', hue = 'Legendary', ax = axi[1][0])
axi[1][0].set_title('Boxplot showing the Comparision between Mega, Legendary, non-mega, Pokemons\nvs.\n Total Power level',fontsize=15)
axi[1][0].set_xlabel('Mega pokemon or Not', fontsize=15)
axi[1][0].set_ylabel('Total Power', fontsize=15)
axi[1][0].tick_params(axis='both', labelsize=10)
axi[1][0].grid(axis = 'y', alpha = 0.4)
axi[1][0].legend(loc='upper left', title='Legendary:-', framealpha=0.4)

sns.barplot(data = dff, x = 'Generation', y = 'Total', hue = 'Legendary', ax=axi[1][1], estimator=np.mean)
axi[1][1].set_title('Bar plot comparing each generation\'s\nmean total power level of Legendary and Non-Legendary pokemons',fontsize=15)
axi[1][1].set_xlabel('Generations', fontsize=15)
axi[1][1].set_ylabel('Total Power', fontsize=15)
axi[1][1].tick_params(axis='both', labelsize=10)
axi[1][1].grid(axis = 'y', alpha = 0.4)
axi[1][1].legend(loc='upper right', title ='Legendary', framealpha = 0.4)
plt.tight_layout(pad= 10)
plt.show() 
fig.savefig('W1-1.png', dpi = 300, bbox_inches = 'tight', pad_inches = 1)


```

### 7. Which is the second fastest non-legendary 'Ghost' type pokemon from 4th generation ?

### 8. How many non-legendary pokemons have stronger defence but weaker attack than Charizard?

```python
#your code here
attack = df.loc['Charizard', ['Attack', 'Defense']].values[0]
defence = df.loc['Charizard', ['Attack', 'Defense']].values[1]
p = (df[(df['Legendary'] == False) & (df['Attack'] < attack)])
q = p[p['Defense'] > defence]
print(f"There are {q.shape[0]} pokemons with attack lower and defence higher than Charizard")
```
